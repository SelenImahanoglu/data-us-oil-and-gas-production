```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gaz verisini yüklüyoruz. 
# Hocanın istediği gibi decimal="," eklemeyi unutmadık!
gas_df = pd.read_csv('data/U.S._natural_gas_production.csv', decimal=",")

# İlk 5 satırı görelim
gas_df.head()
```

```python
Tarih Dönüşümü (İkinci Görev)
```

```python
# Month kolonunu datetime tipine dönüştürüyoruz
gas_df['Month'] = pd.to_datetime(gas_df['Month'])

# Kontrol edelim (dtype: datetime64[ns] görmelisin)
gas_df.info()
```

```python
Yıllara Göre Gruplama (yearly_gas_df)
```

```python
# 'numeric_only=True' ekleyerek sadece sayıları toplamasını sağlıyoruz
yearly_gas_df = gas_df.groupby(gas_df['Month'].dt.year).sum(numeric_only=True)

# Sonucu görelim
yearly_gas_df.head()
```

```python
Tam Yılları Filtreleme
```

```python
# Her yılda kaç satır (ay) olduğunu sayıyoruz ve 12 olanları buluyoruz
full_years_mask = gas_df.groupby(gas_df['Month'].dt.year).size() == 12

# Bu maskeyi kullanarak sadece tam yılları çekiyoruz
filtered_yearly_gas_df = yearly_gas_df[full_years_mask]

filtered_yearly_gas_df
```

```python
Gaz Üretimi Grafiğini Çizelim
```

```python
# ABD toplam gaz üretimini (U.S. kolonu) bar grafiği olarak çiziyoruz
filtered_yearly_gas_df['U.S.'].plot(kind='bar', figsize=(10, 6))

plt.title('Yıllık ABD Toplam Doğal Gaz Üretimi')
plt.ylabel('Üretim Miktarı')
plt.xlabel('Yıl')
plt.show()
```

```python
Petrol Verisine Giriş (oil_df)
```

```python
# Petrol verisini yüklüyoruz. 
# Bu sefer 'parse_dates' ile Month kolonunu en baştan tarihe çevirebiliriz.
oil_df = pd.read_csv('data/U.S._crude_oil_production.csv', parse_dates=['Month'])

# Kolon isimlerindeki olası boşlukları temizleyelim (Hocanın istediği bir detay)
oil_df.columns = oil_df.columns.str.strip()

# İlk bakış
oil_df.head()
```

```python
Yıllık Petrol Verisi ve Filtreleme
```

```python
# Yıllık gruplama
yearly_oil_df = oil_df.groupby(oil_df['Month'].dt.year).sum(numeric_only=True)

# Sadece tam (12 ay) yılları filtreleyelim
full_oil_years = oil_df.groupby(oil_df['Month'].dt.year).size() == 12
filtered_yearly_oil_df = yearly_oil_df[full_oil_years]

# Kontrol
filtered_yearly_oil_df.head()
```

```python
Verileri Birleştirme ve Karşılaştırma
```

```python
# 1. Sadece ABD toplam sütunlarını seçip yeni isimler veriyoruz
# 'U.S. Crude Oil ' yerine 'U.S. Crude Oil' yazıyoruz (Sondaki boşluğu sildik!)
total_gas = filtered_yearly_gas_df[['U.S.']].rename(columns={'U.S.': 'Gas'})
total_oil = filtered_yearly_oil_df[['U.S. Crude Oil']].rename(columns={'U.S. Crude Oil': 'Crude Oil'})

# 2. İki tabloyu yan yana (axis=1) birleştiriyoruz
merged_df = pd.concat([total_gas, total_oil], axis=1)

# 3. Sonuçları kontrol edelim ve grafiği çizelim
merged_df.plot(kind='bar', figsize=(12, 7))
plt.title('ABD Petrol ve Gaz Üretimi Karşılaştırması')
plt.ylabel('Üretim Miktarı')
plt.show()
```
